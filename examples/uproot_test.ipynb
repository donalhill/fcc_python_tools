```python
import uproot4 as uproot
from fcc_python_tools.locations import loc
import matplotlib.pyplot as plt
import awkward1 as ak

file = uproot.open(f"{loc.DATA}/FCCDelphesOutput.root")
tree = file['events']
tree.keys()
```

```python
vars = tree.keys(filter_name="genParticles/genParticles.core*")
vars
#Removing the genParticles/ prefix, and passing the result to arrays(), works
vars_new = []
for v in vars:
    v_new = v.replace('genParticles/','')
    vars_new.append(v_new)
vars_new
```

```python
x = tree.arrays(vars_new, how="zip")
x
```

```python
plt.hist(ak.flatten(x['genParticles.core','p4.px']),bins=100)
plt.show()
```
